# 2. The Pearson correlation coefficient which measures the linear relationship between two datasets

```python
import time
start_time=time.time()
import numpy as np
from scipy import stats
a = np.array([1, 2, 3, 4, 5, 6, 7])
b = np.array([7, 6, 5, 4, 3, 2, 1]) #try [1,2,3,4,5,6,7]
print(stats.pearsonr(a, b)[0])
print("Execution time ",time.time()-start_time)
```

## Create 2 arrays x and y of 25 numbers. Calculate slope and intercept using linear least-squares regression. For a new x and y value predict the output.

### Adjust x and y to print the following output

```python
#to get array using the formula y=mx+c
#where m, x, c will be given

def Array(x,m,c):

  y=[]
  for i in x:
    y1=m*i+c #y=mx+c
    y.append(y1)
  return np.array(y)
```

i) Slope=1, Intercept=0

```python
arr = np.arange(100)
x= np.random.choice(arr, size =25)
y=Array(x,1,0)

print("x: ",x)
print("y: ",y)
print("Slope, intercept: ",stats.linregress(x, y)[0:2])
```

ii) Slope=0, intercept=1

```python
arr = np.arange(100)
x= np.random.choice(arr, size =25)
y=Array(x,0,1)

print("x: ",x)
print("y: ",y)
print("Slope, intercept: ",stats.linregress(x, y)[0:2])
```

# 5. Simple Plot

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10)
y=x
fig, ax = plt.subplots()
ax.plot(x, y,label="$y=x$")
y=2*x
ax.plot(x, y,label="$y=2*$x")    
#Refer https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
#for more options
y=x*x
ax.plot(x, y,label="$y=x^2$") 
ax.legend()
ax.set(xlabel='X', ylabel='Y', title='')
ax.grid()
plt.show()
```

### Simple Plot time complexity n,logn,n*n, n^3, 2^n

Label x and y axis, Give legends, Give title

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

n=np.arange(6)
fig, ax = plt.subplots()

y=n
ax.plot(n, y,label="$y=x$")

y=np.log(n)
ax.plot(n, y,label="$y=logn$")    

y=n*n
ax.plot(n, y,label="$y=n^2$") 

y=n**3
ax.plot(n, y,label="$y=n^3$")

y=2**n
ax.plot(n, y,label="$y=2^n$")

ax.legend
ax.set(xlabel='X', ylabel='Y', title='TIME COMPLEXITIES')
ax.grid()
plt.show()
```
